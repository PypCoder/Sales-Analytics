```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
df.dtypes
```

```python
df['Profit Margin']=df['Profit']/df['Sales']
df.reset_index(inplace=True)
for i in range(0,len(df)):
    df.loc[i,'Row ID']=i+1
```

```python
profit_margin_agg_category=df.groupby('Category')['Profit Margin'].sum()
```

```python
profit_margin_agg_sub_category=df.groupby('Sub-Category')['Profit Margin'].sum()
```

```python
top_products_by_sales=df.groupby('Product Name')['Sales'].sum().nlargest(10)
```

```python
top_customers_by_sales=df.groupby('Customer ID')['Sales'].sum().nlargest(10)
```

```python
least_products_by_sales=df.groupby('Product Name')['Sales'].sum().nsmallest(10)
```

```python
least_customers_by_sales=df.groupby('Customer ID')['Sales'].sum().nsmallest(10)
```

```python
monthly_sales=df.set_index("Order Date").resample('ME')['Sales'].sum()
monthly_sales.index=monthly_sales.index.strftime('%B %Y')
monthly_sales_df=monthly_sales.reset_index()
monthly_sales_df.columns=['Month','Sales']
```

```python
total_sales=df['Sales'].sum()
```

```python
top_customers_by_profit=df.groupby('Customer ID')['Profit'].sum().nlargest(10)
```

```python
least_customers_by_profit=df.groupby('Customer ID')['Profit'].sum().nsmallest(10)
```

```python
top_products_by_profit=df.groupby('Product Name')['Profit'].sum().nlargest(10)
```

```python
least_products_by_profit=df.groupby('Product Name')['Profit'].sum().nsmallest(10)
```

```python
monthly_profits=df.set_index('Order Date').resample('ME')['Profit'].sum()
monthly_profits.index=monthly_profits.index.strftime('%B %Y')
monthly_profits_df=monthly_profits.reset_index()
monthly_profits_df.columns=['Month','Profits']
```

```python
total_profit=df['Profit'].sum()
```

```python
state_summary = df.groupby('State').agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values(by='Sales', ascending=False)
```

```python
pivot_table=df.pivot_table(values='Sales',index='Sub-Category',columns='Category',aggfunc='sum')
```

```python
def reset_df_index(graph_data):
    df2 = pd.DataFrame(graph_data)
    df2 = df2.reset_index()
    df2.columns = df2.columns
    return df2
plot_graph_call_count=1
def plot_graph(graph_data='',graph_type='',color=''):
    global plot_graph_call_count
    plot_graph_call_count += 1
    if graph_data is monthly_sales_df:
        plt.figure(figsize=(8,12))
        plt.scatter(graph_data['Sales'],graph_data['Month'],color=['red' if val<0 else 'green' for val in monthly_sales])
        print(monthly_sales_df)
        df2=monthly_sales_df
    elif graph_data is monthly_profits_df:
        plt.figure(figsize=(8,12))
        plt.scatter(monthly_profits_df['Profits'],monthly_profits_df['Month'],color=['red' if val<0 else 'green' for val in monthly_profits])
        print(monthly_profits_df)
        df2=monthly_profits_df
    else :
        df2=reset_df_index(graph_data)
        print(df2)
        plt.figure(figsize=(6,6))
        graph_data.plot(kind=graph_type,color=color)
        plt.legend()
    plt.savefig('plot.png')
    plt.show()
    sheet_name2 = f'Sheet{plot_graph_call_count}'
    with pd.ExcelWriter("Sorted_Superstore_extracted_data.xlsx",mode='a',engine='openpyxl') as f:
        df2.to_excel(f,sheet_name=sheet_name2,index=True)
    img = Image('plot.png')
    wb=load_workbook("Sorted_Superstore_extracted_data.xlsx")
    ws=wb[sheet_name2]
    ws.add_image(img,"k5")
    wb.save('Sorted_Superstore_extracted_data.xlsx')
```

```python
plot_graph(profit_margin_agg_category,'pie',color='black')
plot_graph(profit_margin_agg_sub_category,'barh',color=['red' if val<0 else 'green' for val in profit_margin_agg_sub_category])
plot_graph(top_products_by_sales,'barh',color='black')
plot_graph(top_products_by_profit,'barh',color='black')
plot_graph(top_customers_by_sales,'barh',color='black')
plot_graph(top_customers_by_profit,'barh',color='black')
plot_graph(least_products_by_sales,'barh',color='black')
plot_graph(least_products_by_profit,'barh',color='black')
plot_graph(least_customers_by_sales,'barh',color='black')
plot_graph(least_customers_by_profit,'barh',color='black')
plot_graph(monthly_sales_df,'scatter','black')
plot_graph(monthly_profits_df,'scatter','black')
```
